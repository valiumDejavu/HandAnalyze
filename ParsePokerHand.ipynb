```python
import re
import codecs


#difine pokerHandClass

class PokerHand:
    position = 'unkonwn'
    cards = ''
    result =''
    board =''


#the file path

pathRead= 'C:\data\onehand.txt'
position=['button' , 'small blind','big blind','UTC','HJ','CutOff' ]

theHand = PokerHand()
#read the txt file

infile=codecs.open(pathRead,'r','utf-8')
oneHandText= infile.readlines()
#print(str(oneHandText))


# get the button seat number
m = re.search('Seat #(\d+) is the button', str(oneHandText))

# get the Hero's seat number
heroName = 'jjpokerfun'
n =re.search('Seat (\d+): '+heroName, str(oneHandText))
positionNumber=0
#calculate the position
if m and n:
    positionNumber = int(n.group()[5])-int(m.group()[6])
    if positionNumber< 0:
        positionNumber=positionNumber+6
    theHand.position = position[positionNumber]
    print(theHand.position)
else:
    print('no seat found')
    




# get the cards
cardsPattern ='Dealt to jjpokerfun \[(.+?)]'
x = re.search(cardsPattern,str(oneHandText))

theHand.cards=x.group(0)[-6]+x.group(0)[-5]+x.group(0)[-3]+x.group(0)[-2]
#print(theHand.cards)


#Gigure out win or lose

#get the summery part
oneHandTextString=str(oneHandText)
summaryPart = oneHandTextString.split('*** SUMMARY ***')
#print(summaryPart[1])


#Figure out win or lose
winPattern1='Seat (\d): (.+?)won'
winPattern2='Seat \d: (.+?)collected'
y = re.search(winPattern1,str(summaryPart[1]))
z = re.search(winPattern2,str(summaryPart[1]))
#print(str(summaryPart[1]))

print(str(y))

if y:
    if(int(y.group(0)[5])== positionNumber):
        print(y.group(0)[5])
        theHand.result='win'
    else:
        theHand.result = 'lose'
elif z:
    if(int(z.group(0)[5])== positionNumber):
        theHand.result = 'win'
    else:
        theHand.result = 'lose'
else:
    theHand.result='lose'

#Figure out Board

boardPattern = 'Board \[(.+?)]'
b=re.search(boardPattern,str(summaryPart[1]))
if b:    
    board=str(b.group(0)[6:])
    print(board)
    theHand.board=board

      
print('')
print('')
print('')
print(theHand.cards,theHand.board,theHand.position,theHand.result)

 #write to the CSV file
    
```
